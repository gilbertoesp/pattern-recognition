```python
import plotly as py
import plotly.graph_objs as go
from IPython.display import display, HTML
from collections import Counter

py.offline.init_notebook_mode(connected=True)
```

```python
# Contabiliza etiquetas
def get_counters(col_name, label):
    full_list = ";".join(col_name)
    each_word = full_list.split(";")
    each_word = Counter(each_word).most_common()
    return pd.DataFrame(each_word, columns = [label, "Participantes"])

# Calcular porcentajes
def percent (row, col):
    count = df[col].count()
    return 100 * row[1] / count
```

```python
def get_fig_pie(data, hole = 0, pull = 0):
    labels = [w if len(w)<=30 else w[0:30]+"..." for w in data.iloc[:,0]]
    dg = [
        go.Pie(labels=labels, 
               values=data.iloc[0:,1],
               hole = hole,
               pull = pull,
               hoverinfo = "label+percent",
              )
    ]
    layout = go.Layout(
        autosize=False,
        width=750,
        height=500,
    )
    return go.Figure(data=dg, layout=layout)

fig = get_fig_pie(data)
py.offline.iplot(fig)
```

```python
def get_fig_bar_mult(columns, labels, x_data, angle_likert = 0):
    colors = random_colors(len(columns))

    traces = []
    for i in range(0, len(x_data[0])):
        for xd, yd in zip(x_data, labels):
            traces.append(go.Bar(
                x=[xd[i]],
                y=[yd],
                orientation='h',
                marker=dict(
                    color=colors[i],
                    line=dict(
                            color='rgb(248, 248, 249)',
                            width=1)
                )
            ))

    layout = go.Layout(
        xaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=False,
            zeroline=False,
            domain=[0.15, 1]
        ),
        yaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=False,
            zeroline=False,
        ),
        barmode='stack',
        paper_bgcolor='rgb(248, 248, 255)',
        plot_bgcolor='rgb(248, 248, 255)',
        margin=dict(
            l=120,
            r=10,
            t=140,
            b=80
        ),
        hovermode='closest',
        showlegend=False,
    )

    annotations = []
    for yd, xd in zip(labels, x_data):
        # labeling the y-axis
        annotations.append(dict(xref='paper', yref='y',
                                x=0.14, y=yd,
                                xanchor='right',
                                text=str(yd),
                                font=dict(family='Arial', size=14,
                                          color='rgb(67, 67, 67)'),
                                showarrow=False, align='right',
                               )
                          )
        # labeling the first Likert scale (on the top)
        if yd == labels[-1]:
            annotations.append(dict(xref='x', yref='paper',
                                    x=xd[0] / 2, y=1.1,
                                    text=top_labels[0],
                                    font=dict(family='Arial', size=14,
                                              color='rgb(67, 67, 67)'),
                                    showarrow=False,
                                    textangle = angle_likert
                                   )
                              )
        space = xd[0]
        for i in range(1, len(xd)):
                # labeling the Likert scale
                if yd == labels[-1]:
                    annotations.append(dict(xref='x', yref='paper',
                                            x=space + (xd[i]/2), y=1.1,
                                            text=top_labels[i],
                                            font=dict(family='Arial', size=14,
                                                      color='rgb(67, 67, 67)'),
                                            showarrow=False,
                                            textangle = angle_likert
                                           )
                                      )
                space += xd[i]
    layout['annotations'] = annotations

    return go.Figure(data=traces, layout=layout)
```

```python
import random

def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCD') for j in range(6)])
                 for i in range(number_of_colors)]
    return color
```

```python
columns = list(df.columns.values[38:43])
labels = ['Llamada telefónica', 'Correo privado', 'Correo de trabajo', 
          'Mensaje al sitio de trabajo','Mensaje en redes sociales']

top_labels = []
x_data = []
counter = 39
for col, label in zip(columns, labels):
    display(HTML('<hr style="border-width: 3px;">'))
    display(HTML('<span style="color:blue; font-weight: bold;">' + str(counter) + ". " + 
                 col + ": " +'</span>' + '<span style="font-weight: bold;">'  + 
                 label +'</span>'))
    print(df[col].describe())
    d = get_counters_numeric(df[col].dropna(), label)
    d['Porcentaje'] = d.apply (lambda row: percent (row, col), axis=1)
    display(HTML(d.to_html()))
    
    dsorted = d.sort_values(d.columns[0])
    if len(top_labels) == 0:
        top_labels = list(dsorted.iloc[:,0])
    x_data.append(list(dsorted.iloc[:,1]))
    counter += 1

fig = get_fig_bar_mult(columns, labels, x_data)
    
py.offline.iplot(fig)
```

Para variables que corresponden a una misma pregunta con tres variantes, como en este caso, es conveniente comparar lado a lado los resultados. 
</div>

```python
def get_fig_bar_group(data, labels, option = 'group'): #'group' / 'stack'
    traces = []
    colors = random_colors(len(data))
    for i in range(len(data)):
        traces.append(go.Bar(
            x=data[i].iloc[:,0],
            y=data[i].iloc[:,1],
            name=labels[i],
            marker=dict(
                color=colors[i]
            )
        ))
        
    layout = go.Layout(
        xaxis=dict(tickangle=15),
        barmode=option,
    )
    
    return go.Figure(data=traces, layout=layout)
```

```python
columns = list(df.columns.values[62:65])
labels = ["Afinidad con otros desarrolladores", "Competidor de otros desarrolladores",
         "Inferior a otros desarrolladores"]

data = []
counter = 63
sorterIndex = {}
for col, label in zip(columns, labels):
    display(HTML('<hr style="border-width: 3px;">'))
    display(HTML('<span style="color:blue; font-weight: bold;">' + str(counter) + ". " + 
                 col + ": " +'</span>' + '<span style="font-weight: bold;">'  + 
                 label +'</span>'))
    print(df[col].describe())

    d = get_counters(df[col].dropna(), label)
    d['Porcentaje'] = d.apply (lambda row: percent (row, col), axis=1)
    display(HTML(d.to_html()))

    if len(sorterIndex) == 0:
        indices = list(d.iloc[:,0])
        sorterIndex = dict(zip(indices, range(len(indices))))
    else:
        # Utilizar el tipo "category" para usarlo como criterio de ordenamiento
        d.iloc[:,0] = d.iloc[:,0].astype("category")
        d.iloc[:,0].cat.set_categories(sorterIndex, inplace=True)
        d = d.sort_values([d.columns[0]])
    counter += 1
        
    data.append(d)
    
fig = get_fig_bar_group(data, labels) 
py.offline.iplot(fig)
```

```python
import squarify

def get_fig_treemap(data):
    x = 0
    y = 0
    width = 50
    height = 50

    data = data.head(25)
    type_list = data.iloc[:,0]
    values = data.iloc[:,1]

    normed = squarify.normalize_sizes(values, width, height)
    rects = squarify.squarify(normed, x, y, width, height)
    
    colors = random_colors(20)
    shapes = []
    annotations = []
    counter = 0
    color = 0

    for r in rects:
        shapes.append( 
            dict(
                type = 'rect', 
                x0 = r['x'], 
                y0 = r['y'], 
                x1 = r['x']+r['dx'], 
                y1 = r['y']+r['dy'],
                line = dict( width = 1 ),
                fillcolor = colors[color]
            ) 
        )
        annotations.append(
            dict(
                x = r['x']+(r['dx']/2),
                y = r['y']+(r['dy']/2),
                text = "{}".format(type_list[counter]),
                showarrow = False
            )
        )
        counter = counter + 1
        color = color + 1
        if color >= len(colors):
            color = 0

        trace0 = go.Scatter(
            x = [ r['x']+(r['dx']/2) for r in rects ],
            y = [ r['y']+(r['dy']/2) for r in rects ],
            text = [ str(v) for v in values ], 
            mode = 'text',
        )

        layout = dict(
            height=600, 
            width=750,
            xaxis=dict(showgrid=False,zeroline=False),
            yaxis=dict(showgrid=False,zeroline=False),
            shapes=shapes,
            annotations=annotations,
            hovermode='closest',
            font=dict(color="#FFFFFF"),
            margin = go.Margin(
                l=0,
                r=0,
                pad=0
            )
        )

    return dict(data=[trace0], layout=layout)

figure = get_fig_treemap(data)
py.offline.iplot(figure)
```

Aunque estos resultados parecen "razonables", a primera vista, podemos esperar que haya errores, debido a la confusión de la terminología. Observemos los resultados sin agrupar:
